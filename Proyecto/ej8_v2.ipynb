```python
import numpy as np
import numpy.linalg as linalg
import matplotlib.pyplot as plt
from numpy import linalg as la

rango = [-3.5, 3.5]

def plot(array, vectors,shape = (10,10), name = False):
    fig, ax = plt.subplots(figsize = shape)
    for v in array:
        ax.plot(v[0], v[1])
    
    start = [0,0]
    count = 1
    color = ["black", "blue", "red", "skyblue"]
    if name == False:
        for v in vectors:
            r = (v[0]**2+v[1]**2)**0.5
            ax.plot([-4*v[0]/r, 4*v[0]/r], [-4*v[1]/r, 4*v[1]/r])
            ax.quiver(*start, v[0], v[1], scale=1, scale_units = 'xy', angles = 'xy')
            ax.annotate(f"$v_{count}$", (v[0], v[1]), color = color[(count-1)%len(color)], fontsize = 20, alpha = 0.8)
            count += 1
    else:
        for v in vectors:
            r = (v[0]**2+v[1]**2)**0.5
            ax.plot([-4*v[0]/r, 4*v[0]/r], [-4*v[1]/r, 4*v[1]/r])
            ax.quiver(*start, v[0], v[1], scale=1, scale_units = 'xy', angles = 'xy')
            ax.annotate(f"$\sigma_{count} u_{count}$", (v[0], v[1]), color = color[(count-1)%len(color)], fontsize = 20, alpha = 0.8)
            count += 1
    
    plt.xlim(rango)
    plt.ylim(rango)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
def unit_ball(resolution = 50):
    theta = np.linspace(-np.pi, np.pi, resolution)
    X = np.cos(theta)
    Y = np.sin(theta)
    return (X, Y)
```

```python
plot([unit_ball()], [(1,0), (0,1)])
```

```python
fig, ax = plt.subplots(figsize = (10,10))

A = np.array([[3, 0],[0, -2]])
U, D, V = np.linalg.svd(A)
v_1 = np.array([[1], [0]])
v_2 = np.array([[0], [1]])
elipse = np.array([ A@np.array([unit_ball()[0][i], unit_ball()[1][i]]) for i in range(50)])
plot([(elipse[:, 0], elipse[:, 1])], [A@v_1, A@v_2], (10,10), False)

u1 = [D[0]*U[0,0], D[0]*U[0,1]]
v1 = [D[1]*U[1,0], D[1]*U[1,1]]
plot([(elipse[:, 0], elipse[:, 1])], [u1, v1], (10,10), True)

plt.xlim(rango)
plt.ylim(rango)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
