# Pregunta 2

### a)

Sea $p\in [1, +\infty)$, $n\in\mathbb{N}$ y $x \in \mathbb{K}^n$.

Notemos lo siguiente:

\begin{align}
& &\Vert x\Vert_\infty^p &\geq \vert x_i \vert^p , \forall i= 1, ..., n \\
&\Rightarrow& \sum_{k=1}^n  \Vert x\Vert_\infty^p &\geq \sum_{k=1}^n \vert x_k \vert^p\\
&\Leftrightarrow & n  \Vert x\Vert_\infty^p &\geq  \Vert x \Vert_p^p\\
&\Rightarrow & n^{\frac{1}{p}}  \Vert x\Vert_\infty &\geq  \Vert x \Vert_p
\end{align}

Por otro lado, fijemos $k_0$ tal que $\Vert x\Vert_\infty = \vert x_{k_0}\vert$,
\begin{align}
&\Rightarrow \Vert x\Vert_\infty^p =\vert x_{k_0} \vert^p \leq \sum_{k=1}^n \vert x_k \vert^p = \Vert x \Vert_p^p\\
&\Rightarrow \Vert x\Vert_\infty \leq \Vert x \Vert_p
\end{align}

Por lo tanto:
\begin{align}
 \Vert x\Vert_\infty \leq \Vert x \Vert_p \leq n^{\frac{1}{p}}  \Vert x\Vert_\infty 
\end{align}

Para la otra desigualdad veamos lo siguiente:

\begin{align}
&&0 &\leq 2\vert a \vert \vert b \vert\\
&\Rightarrow& a^2+b^2 &\leq a^2 + 2\vert a \vert \vert b \vert +b^2\\
&\Rightarrow& \sqrt{a^2+b^2} &\leq \vert a \vert + \vert b \vert
\end{align}

Con este resultado previo, demostraremos por inducción: 
\begin{align}
P(n) : \sqrt{\sum_{k=1}^n x_k^2} \leq \sum_{k=1}^n \vert x_k \vert
\end{align}

Claramente $P(0)$ se satisface:
\begin{align}
P(1) : \sqrt{\sum_{k=1}^1 x_k^2} = \vert x_1 \vert \leq \vert x_1 \vert = \sum_{k=1}^1 \vert x_k \vert
\end{align}

Con esto, suponemos $P(n)$ válido para algún $n\in \mathbb{N}$ y buscamos provar que

\begin{align}
P(n+1) : \sqrt{\sum_{k=1}^{n+1} x_k^2} \leq \sum_{k=1}^{n+1} \vert x_k \vert, \text{ se satisface}
\end{align}

Reescribamos el lado derecho utilizando el resultado previo:

\begin{align}
\sqrt{\sum_{k=1}^{n+1} x_k^2} = \sqrt{\sum_{k=1}^{n} x_k^2 + x_{n+1}^2} &\leq \sqrt{\sum_{k=1}^{n} x_k^2} + \vert x_{n+1} \vert, \text{ luego por hipótesis de inducción}\\
&\leq \sum_{k=1}^{n} \vert x_k \vert + \vert x_{n+1} \vert = \sum_{k=1}^{n+1} \vert x_k \vert
\end{align}
Demostrando por inducción lo propuesto.

Notemos que hemos demostrado:

\begin{align}
\Vert x \Vert_2 = \sqrt{\sum_{k=1}^{n+1} x_k^2} \leq \sum_{k=1}^{n+1} \vert x_k \vert = \Vert x \Vert_1
\end{align}

Por otro lado, primero veamos un resultado previo por inducción:

\begin{align}
P(n) : \sum_{j=2}^n \sum_{k=1}^{j-1} \vert x_j \vert \vert x_k \vert = \sum_{j=1}^{n-1} \sum_{k=j+1}^{n} \vert x_j \vert \vert x_k \vert
\end{align}

Notemos que efectivamente se tiene $P(2)$:

\begin{align}
P(2) : \sum_{j=2}^2 \sum_{k=1}^{j-1} \vert x_j \vert \vert x_k \vert = \vert x_2 \vert \vert x_1 \vert = \vert x_1 \vert \vert x_2 \vert = \sum_{j=1}^{1} \sum_{k=j+1}^{2} \vert x_j \vert \vert x_k \vert
\end{align}

Ahora supongamos que se tiene $P(n)$ para algún $n\in\mathbb{N}$, demostraremos que también se tiene:

\begin{align}
P(n+1) : \sum_{j=2}^{n+1} \sum_{k=1}^{j-1} \vert x_j \vert \vert x_k \vert = \sum_{j=1}^{n} \sum_{k=j+1}^{n+1} \vert x_j \vert \vert x_k \vert
\end{align}

Reescribamos el lado izquierdo:
\begin{align}
\sum_{j=2}^{n+1} \sum_{k=1}^{j-1} \vert x_j \vert \vert x_k \vert &= \sum_{j=2}^{n} \sum_{k=1}^{j-1} \vert x_j \vert \vert x_k \vert + \sum_{k=1}^{n} \vert x_{n+1} \vert \vert x_k \vert \text{ , ahora por hipótesis de inducción }\\
&=\sum_{j=1}^{n-1} \sum_{k=j+1}^{n} \vert x_j \vert \vert x_k \vert + \sum_{k=1}^{n} \vert x_{n+1} \vert \vert x_k \vert\\
&=\sum_{j=1}^{n-1} \sum_{k=j+1}^{n} \vert x_j \vert \vert x_k \vert + \sum_{j=1}^{n-1} \vert x_{n+1} \vert \vert x_j \vert + \vert x_{n+1} \vert \vert x_n \vert\\
&=\sum_{j=1}^{n-1} \left(\sum_{k=j+1}^{n} \vert x_j \vert \vert x_k \vert + \vert x_{n+1} \vert \vert x_j \vert \right)  + \vert x_{n+1} \vert \vert x_n \vert\\
&=\sum_{j=1}^{n-1} \left(\sum_{k=j+1}^{n} \vert x_j \vert \vert x_k \vert + \vert x_{n+1} \vert \vert x_j \vert \right)  + \sum_{j = n}^{n} \vert x_{n+1} \vert \vert x_j \vert\\
&=\sum_{j=1}^{n} \left(\sum_{k=j+1}^{n} \vert x_j \vert \vert x_k \vert + \vert x_{n+1} \vert \vert x_j \vert \right)\\
&=\sum_{j=1}^{n} \sum_{k=j+1}^{n+1} \vert x_j \vert \vert x_k \vert 
\end{align}

Demostrando así, lo propuesto.

Con esto podemos reescribir lo siguiente:
\begin{align}
 \left( \sum_{j=1}^n \vert x_j \vert \right)^2 &= \sum_{j=1}^{n} \vert x_j\vert \sum_{k=1}^{n} \vert x_k \vert\\
&= \sum_{j=1}^{n}\sum_{k=1}^{n} \vert x_j\vert  \vert x_k \vert\\
&= \sum_{j=1}^{n}\left(\sum_{k=1}^{j-1}  \vert x_j\vert  \vert x_k \vert + \sum_{k=j}^{n}  \vert x_j\vert  \vert x_k \vert \right)\\
&= \sum_{j=2}^{n}\sum_{k=1}^{j-1}  \vert x_j\vert  \vert x_k \vert + \sum_{j=1}^{n}\sum_{k=j}^{n}  \vert x_j\vert  \vert x_k \vert\\
&= \sum_{j=2}^{n}\sum_{k=1}^{j-1}  \vert x_j\vert  \vert x_k \vert + \sum_{j=1}^{n-1}\sum_{k=j+1}^{n}  \vert x_j\vert  \vert x_k \vert + \sum_{j=1}^{n} x_j^2\\
\end{align}

Por lo anterior
\begin{align}
= 2\sum_{j=2}^{n}\sum_{k=1}^{j-1}  \vert x_j\vert  \vert x_k \vert + \sum_{j=1}^{n} x_j^2\\
\end{align}

Así, con todo esto estamos listos para demostrar la última desigualdad que nos queda:

Notemos lo siguiente:

\begin{align}
&&0 &\leq \sum_{j=1}^{n} \sum_{k=1}^{j} (\vert x_j \vert - \vert x_k \vert)^2\\
&\Leftrightarrow & 0 &\leq \sum_{j=1}^{n} \sum_{k=1}^{j} \left( x_j^2 + x_k^2- 2 \vert x_j \vert \vert x_k \vert \right)\\
&\Leftrightarrow & 0 &\leq \sum_{j=1}^{n} \left( jx_j^2 + \sum_{k=1}^{j}x_k^2- 2 \sum_{k=1}^{j}\vert x_j \vert \vert x_k \vert \right)\\
\end{align}

Notemos lo siguiente:

Para $\alpha \in {1, ..., n}$, se tiene que $x_\alpha$ solo aparece para $ j \geq \alpha$ en la sumatoria $\sum_{k=1}^{j}x_k^2$ entonces $x_\alpha$ aparece $(n+1-\alpha)$ veces. Luego cuando $j$ vale $\alpha$, por $jx_j$, $x_\alpha$ aparece otras $\alpha$ veces. Luego en total $x_\alpha$ aparece $(n+1-\alpha) + \alpha = n+1$ veces. Así obtenemos lo que sigue:

\begin{align}
&& 0 &\leq \sum_{j=1}^{n} \left( jx_j^2 + \sum_{k=1}^{j}x_k^2- 2 \sum_{k=1}^{j}\vert x_j \vert \vert x_k \vert \right)\\
&\Leftrightarrow & 0 &\leq (n+1) \sum_{j=1}^n x_j^2 -2 \sum_{j=1}^{n}  \sum_{k=1}^{j}\vert x_j \vert \vert x_k \vert \\
&\Leftrightarrow & 0 &\leq (n+1) \sum_{j=1}^n x_j^2 -2 \sum_{j=1}^{n}  \left(\sum_{k=1}^{j-1}\vert x_j \vert \vert x_k \vert +  x_j^2\right)\\
&\Leftrightarrow & 0 &\leq (n+1) \sum_{j=1}^n x_j^2 -2 \sum_{j=2}^{n}  \sum_{k=1}^{j-1}\vert x_j \vert \vert x_k \vert -2\sum_{j=1}^{n}x_j^2\\
&\Leftrightarrow & \sum_{j=1}^{n}x_j^2 +2 \sum_{j=2}^{n}  \sum_{k=1}^{j-1}\vert x_j \vert \vert x_k \vert  &\leq n \sum_{j=1}^n x_j^2 \\
&\Leftrightarrow & \sum_{j=1}^{n} \vert x_j\vert \sum_{k=1}^{n} \vert x_k \vert &\leq n \sum_{j=1}^n x_j^2 \\
&\Leftrightarrow & \left(\sum_{j=1}^{n} \vert x_j \vert \right)^2 &\leq n \sum_{j=1}^n x_j^2 \\
&\Leftrightarrow & \Vert x\Vert_1 = \sum_{j=1}^{n} \vert x_j \vert &\leq \sqrt{n} \sqrt{\sum_{j=1}^n x_j^2} = \sqrt{n} \Vert x \Vert_2\\
\end{align}

Concluimos así:
\begin{equation}
\Vert x \Vert_2\leq \Vert x\Vert_1 \leq \sqrt{n} \Vert x \Vert_2\\
\end{equation}

### b)

Sea $A\in \mathbb{K}^{m \times n}$, luego

\begin{align}
\Vert A \Vert_\infty = 
\sup_{x \in \mathbb{K}^n\\ \Vert x \Vert_\infty = 1}  \Vert Ax \Vert_\infty = 
\sup_{x \in \mathbb{K}^n\\ \Vert x \Vert_\infty = 1} \max_{i = 1, ..., m} \left\vert \sum_{j = 1}^n a_{ij} x_j \right\vert \leq
\sup_{x \in \mathbb{K}^n\\ \Vert x \Vert_\infty = 1} \max_{i = 1, ..., m}  \sum_{j = 1}^n \left\vert a_{ij} x_j \right\vert =
\max_{i = 1, ..., m}  \sum_{j = 1}^n \left\vert a_{ij}\right\vert
\end{align}

Por otro lado, sea $i_0 \in \{1, ..., m\}$ tal que 

\begin{align}
\max_{i = 1,...,m} \sum_{j = 1}^n \vert a_{ij} \vert = \sum_{j = 1}^n \vert a_{i_0 j} \vert
\end{align}

Entonces podemos definir $x \in \mathbb{K}^{n}$ dado por

\begin{align}
x_j = sign(a_{i_0 j}) , j \in \{ 1,...,n\}
\end{align}

De cuya definición concluimos que $\Vert x \Vert_\infty = 1$, entonces

\begin{align}
\max_{i = 1, ..., m}  \sum_{j = 1}^n \left\vert a_{ij}\right\vert =
\sum_{j = 1}^n \vert a_{i_0 j} \vert =
\left\vert\sum_{j = 1}^n  a_{i_0 j} x_j \right\vert \leq
\sup_{x \in \mathbb{K}^n\\ \Vert x \Vert_\infty = 1} \max_{i = 1, ..., m} \left\vert \sum_{j = 1}^n a_{ij} x_j \right\vert
\end{align}

Concluyendo así lo que sigue

\begin{align}
\Vert A \Vert_\infty = \max_{i = 1, ..., m}  \sum_{j = 1}^n \left\vert a_{ij}\right\vert
\end{align}
