Primero, se definir치 el metodo de las potencias, el cual se utilizar치 paraobtener el valor y vector dominante de alguna matriz dada.

```python
import numpy as np

def my_pow(A,x,tol,N):
    
    lmb = 0. # lambda inicial
    k = 0 # contador de iteraciones
    err = 1. # error inicial (tomar valor grande)
    np.set_printoptions(suppress=True,precision=1000000000)
    while k<=N and err>tol:
        y = A@x.astype(float)
        # normaliza y
        indx=np.argmax(np.abs(y))
        c1 = y[indx,0].astype(float)
        dc = np.abs(lmb-c1)# distancia entre lambda^{(k)} y lambda^{(k+1)}
        y = (1/c1)*y
        # actualiza x y lambda (check para convergencia)
        dv = np.linalg.norm(x-y) # distancia entre V^{(k)} y V^{(k+1)}
        err = np.max([dc,dv]) # se toma el maximo de los errores
        x = y
        lmb=c1
        k+=1
    v=x.astype(float)
    
    return lmb,v,err,k
```

Junto con ello, se escribira la funci칩n que generara la matriz pedida en la pregunta 14 inciso b.

```python
def generate_matriz(n):
    Matriz = np.zeros((n,n),dtype=float)
    for i in range(0,n):
        for j in range(0,n):
            if i==j:
                Matriz[i,j]=(i+1)+(j+1)
            else:
                Matriz[i,j]=(i+1)*(j+1)
    return Matriz
Matriz=generate_matriz(15)
print(Matriz)
```

Finalmente, se tiene el metodo de deflaci칩n Wielandt, el cual mediante la subrutina del metodo de las potencias, se tendran todos los valores y vectores propios.

```python
def Deflation(A):
    np.set_printoptions(suppress=False,precision=1000000000)
    x0 = np.ones((A.shape[0],1))
    lamb={}
    vec={}
    wec={}
    err={}
    k={}
    x={}
    Matrix={}
    for i in range(1,A.shape[0]+1):
        if i==1:
            Matrix[i]=A
            lamb[i], vec[i], err[i], k[i] = my_pow(Matrix[i],x0,1E-10,100)
            print(lamb[i],"\n\n", vec[i],"\n\n")
            v=vec[i]
            M=Matrix[i]
            wec[i]=vec[i]
        else:
            max=0
            for j in range(0,A.shape[0]-1):
                if v[j]>max: 
                    x[i] = M[j].transpose()/(lamb[i-1]*v[j])
            x[i].shape=(A.shape[0],1)
            Matrix[i] = Matrix[i-1]-lamb[i-1]*wec[i-1]@x[i].transpose()
            lamb[i], vec[i], err[i], k[i] = my_pow(Matrix[i],x0,1E-10,100)
            wec[i] = vec[i]
            c=1
            d=0
            while c<i:
                vec[i]=(lamb[i]-lamb[i-c])*vec[i]+lamb[i-c]*wec[i-c]*(x[i-d].transpose()@vec[i])
                c=c+1
                d=d+1
            c=1
            d=0
            if i!=A.shape[0]:
                v=wec[i]
                M=Matrix[i]
            print(lamb[i],"\n\n", vec[i],"\n\n")
```
